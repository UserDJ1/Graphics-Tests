```julia
using Plots
```

```julia
meshgrid(x, y) = (repeat(x, outer=length(y)), repeat(y, inner=length(x)))
```

```julia
x, y = meshgrid(0:0.2:2, 0:0.2:2)
u = @. cos(x) * y
v = @. sin(x) * y
```

```julia
scale = 0.2
u = @. scale * cos(x) * y
v = @. scale * sin(x) * y
```

```julia
quiver(x, y, quiver=(u, v))
```
